```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


def generation_and_plot(m):
    
    n = 10000
    X1 = np.random.normal(loc = 0, size = (n, 2))
    X2 = np.random.normal(loc = m, size = (n, 2))
    X = np.vstack([X1, X2])
    y = np.array([0]*n + [1]*n)
    plt.tight_layout()
    sns.scatterplot(x = X[:,0], y = X[:,1], hue = y)
    
    return(X, y)

X3, y3 = generation_and_plot(10)
generation_and_plot(10)
```

```python
def svc_classification(X, y):

    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 5)

    model = SVC(kernel = 'linear', class_weight = 'balanced')
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    
svc_classification(X3, y3)
```

```python
X4, y4 = generation_and_plot(0)
generation_and_plot(0)
svc_classification(X4, y4)
```
